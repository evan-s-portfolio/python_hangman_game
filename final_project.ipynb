## Introduction

For the final project, I created a hangman game. The aim of the project is to create a simple game that demonstrates my ability to program in Python. Also, playing hangman is fun! 

Note that because the game is simple and there isn't a lot of code (about 125 lines in the main program, although the actual amount of code is much less becuase I put in lots of comments), I decided not to split it up into functions or create a class. Instead, there is just one function and at the bottom of the main program the function is called to start the game.

## List of Files

The project contains the following files:
1. final_project.ipynb: A Jupyter notebook with info on the project.
1. hangman_game.py: A python file that contains the program code for the hangman game.
1. hangman_images.py: A python file that contains text art images used by the game.
1. word_list.py: A python file that contains a list of words used by the game.

## How to Run the Project

To play the hangman game:
1. Create a new folder and save all the project files into the folder.
1. Start VS Code and open the folder where the files are saved.
1. Open the hangman_game.py file and click on the Run button to run it in a terminal.
1. Have fun! (Warning: The game may be addictive!)
